```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
```

```python
def im(grid, title = '', origin = 'lower'):
    plt.imshow(grid, origin=origin)
    plt.colorbar()
    plt.title(title)
    plt.show()
```

```python
def domain(M):    
    x = np.linspace(0, 1, M)
    y = np.linspace(0, 1, M)
    xx, yy = np.meshgrid(x, y)
    domain = np.zeros((M, M))
    domain = yy < 1 - xx**2
    return domain


def domain2(dom_mx):
    vec = np.zeros(len(dom_mx[0,:]))
    for i in range(len(vec)):
        vec[i] = np.sum(dom_mx[i])
    vec = np.flip(vec).astype(int)
    return vec
M=5
print(domain2(domain(M)[::-1]))
```
